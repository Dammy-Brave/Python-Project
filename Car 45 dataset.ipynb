### Data Cleaning
* Removing irrelevant column (Unnamed: 0)
* Handling null values
* Splitting the region

```python
# dropping irrelevant columns
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
#checking null values
df.isnull().sum()
```

```python
#handling missing values
#dropping columns with small non-null values
df2 = df.drop(columns = ['trim', 'drive_train', 'reg_city', 'seat', 'num_cylinder', 'horse_power', 'body_build'])
```

```python
df2.head()
```

```python
#splitting the region column
df2[['State', 'City']] = df2['region'].str.split(',', expand = True)
```

```python
df2.isnull().sum()
```

```python
#drop null values
df2 = df2.dropna()
```

```python
df2.info()
```

### Analysis Questions

Question 1: Average car price by States

```python
df2[['amount','State']]
```

```python
avg_price_by_state = df2['amount'].groupby(df2['State']).mean().reset_index()
```

```python
avg_price_by_state 
```

```python
# visaulizing avg_price_by_state
sns.barplot(data = avg_price_by_state, x = 'State', y = 'amount')
plt.title('Average Car price by States')
plt.ylabel('Average price')
plt.xlabel('State')
plt.xticks(rotation = 60);
```

 Question 2: Average Car price by make

```python
avg_price_by_make = df2['amount'].groupby(df2['make']).mean().reset_index().sort_values(by = 'amount', ascending = False)[0:10]
```

```python
avg_price_by_make
```

```python
# visaulizing avg_price_by_state
sns.barplot(data = avg_price_by_make, x = 'make', y = 'amount')
plt.title('Average Car price by Make')
plt.ylabel('Average price')
plt.xlabel('make')
plt.xticks(rotation = 60);
```

Question 3: Average car price by Transmission

```python
avg_price_by_transmission = df2['amount'].groupby(df2['transmission']).mean().reset_index().sort_values(by = 'amount')
```

```python
avg_price_by_transmission 
```

```python
# visaulizing avg_price_by_transmission
sns.barplot(data = avg_price_by_transmission, x = 'transmission', y = 'amount')
plt.title('Average Car price by Transmission')
plt.ylabel('Average price')
plt.xlabel('Transmission');
```

Question 4: Sales Trend Over Year

```python
sales_trend_over_year = df2['amount'].groupby(df2['year_of_man']).mean().reset_index().sort_values(by = 'amount')
```

```python
sales_trend_over_year
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data = sales_trend_over_year, x='year_of_man', y='amount')
plt.title('Sales Trend Over Year')
plt.ylabel('Average price')
plt.xlabel('Year of Make');
plt.xticks(rotation = 60);
```

### Insight
* The average cost of an automobile is highest in Anambra State, with Lagos State, Ondo State, and Abuja (FCT) following closely after. This implies that, in comparison to other states, Anambra State has the greatest purchasing power or a greater taste for luxury cars.

###  Recommendation
* To better fulfill local demand, adjust marketing and sales methods based on region preferences and economic realities.
* To promote the automobile industry's sustained expansion, invest in infrastructure such as road networks and auto dealerships in states with high car sales.

### Insight:

### Insight :
* At 5,077,071 units sold, Automatic transmissions (AT) have the highest sales volume, indicating a significant consumer preference for usability and convenience.
* Continuously Variable Transmission (CVT) has Second highest with 4,922,647 units growing popularity due to fuel efficiency and smooth driving  experience.
* Automated Manual Transmission (AMT) has the Third with 4,702,500 units balances between manual and automatic features, appealing to budget-conscious    buyers who seek convenience.
* With 2,743,546 units sold, manual transmissions are the least popular. This declining preference is probably brought on by people's inclination for simpler and more convenient driving experiences.

### Recommendation

* Market and Promote: Emphasize the affordability of AMT as well as the ease of use of AT, CVT, and other options.
* Build Products: Put an emphasis on optimizing CVT and AT models, as well as boosting AMT efficiency.
* Educate Customers: Offer details on the advantages of transmissions and promote test drives.
* Handle Support and Inventory: Provide robust maintenance plans, keep an eye on trends, and stock more AT and CVT vehicles.
